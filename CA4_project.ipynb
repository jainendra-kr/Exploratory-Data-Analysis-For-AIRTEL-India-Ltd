```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
My_data= pd.read_csv('C:\\Users\\jaine\\Desktop\\LPU classwork\\3rd sem\\Assignement\\CAP 776\\Project\\Airtel_Prepaid.csv')
My_data. rename(columns = {'Additional Benefits':'Airtal Pack Benefits'}, inplace = True)
My_data.head()
```

```python
My_data['Data(GB)'].unique()
```

```python
features = pd.get_dummies(My_data['Airtal Pack Benefits'])
features.head()
```

```python
frames=[features,My_data]
New_data= pd.concat(frames,axis=1)
New_data. rename(columns = {'0':'No_Benefits'}, inplace = True)
New_data.head()
```

```python
f, ax = plt.subplots(figsize=(15,10))
mask = np.triu(np.ones_like(New_data.corr(), dtype=bool))
cmap = sns.diverging_palette(500, 15, as_cmap=True)
sns.heatmap(New_data.corr(), annot=True, mask = mask, cmap=cmap)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
My_data['Airtal Pack Benefits'] = le.fit_transform(My_data['Airtal Pack Benefits'])
My_data.head()
```

```python
f, ax = plt.subplots(figsize=(7, 3))
mask = np.triu(np.ones_like(My_data.corr(), dtype=bool))
cmap = sns.diverging_palette(500, 15, as_cmap=True)
sns.heatmap(My_data.corr(), annot=True, mask = mask, cmap=cmap)
```
